```python
# necesary imports
import numpy
from numpy import array
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.datasets import imdb
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import LSTM, Dropout, Bidirectional
from tensorflow.keras.layers import Embedding 
from tensorflow.keras.preprocessing import sequence
from tensorflow.keras.models import load_model
import re
import preprocess_js
import numpy as np
from nltk.tokenize import word_tokenize
import nltk

numpy.random.seed(7)
```

```python
top_words = 20000
(X_train,y_train), (X_test,y_test) = imdb.load_data(num_words = top_words)
```

```python
word2id = imdb.get_word_index()
id2word = {}
for i,j in word2id.items():
    id2word[j]=i
```

```python
# Max review len and min review len

print("max review length",len(max(X_test+X_train,key=len)))
print("min review length",len(min(X_test+X_train,key=len)))
```

```python
"""Now that we have the data but we can't feed this unless all the data is of equal length
   so we shall truncate data exceeding our limit and pad data under out fixed limit"""

max_review_length = 800
X_train = sequence.pad_sequences(X_train,maxlen=max_review_length)
X_test = sequence.pad_sequences(X_test,maxlen = max_review_length)
```

```python
embedding_vector_length = 50
# model = Sequential()
# model.add(Embedding(top_words, embedding_vector_length,input_length=max_review_length))
# model.add(Dropout(0.2))
# model.add(LSTM(100))
# model.add(Dropout(0.2))
# model.add(Bidirectional(LSTM(100)))

inputs = keras.Input(shape=(None,), dtype="int32")
x = Embedding(top_words, embedding_vector_length,input_length=max_review_length)(inputs)
x = Dropout(0.2)(x)
x = Bidirectional(LSTM(64, return_sequences=True))(x)
x = Dropout(0.2)(x)
x = Bidirectional(LSTM(64))(x)
x = Dropout(0.2)(x)
outputs = Dense(1, activation="sigmoid")(x)
model = keras.Model(inputs,outputs)
model.compile("adam", "binary_crossentropy", metrics=["accuracy"])
model.fit(X_train, y_train, batch_size=64, epochs=30, validation_data=(X_test, y_test))
```

```python
print(model.summary())
```

```python
model.save('my_model.h5')
```

```python
scores = model.evaluate(X_test, y_test, verbose=0)
print("Accuracy: %.2f%%" % (scores[1]*100))
```
